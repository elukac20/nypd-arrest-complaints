```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
complaints = pd.read_csv("NYPD_Complaint_Data_Historic_20250530.csv")
```

```python
arrests = pd.read_csv("NYPD_Arrests_Data__Historic__20250530.csv")
```

```python
complaints_sorted = complaints.sort_values(by=['CMPLNT_FR_DT','CMPLNT_FR_TM'])
complaints_sorted
```

```python
complaints_sorted['CMPLNT_FR_DT']=pd.to_datetime(complaints_sorted['CMPLNT_FR_DT'], errors='coerce')
complaints_sorted['CMPLNT_FR_DT'].dropna(inplace=True)
```

```python
complaints_sorted['CMPLNT_FR_DT']
```

```python
morethandate = complaints_sorted['CMPLNT_FR_DT'] >= pd.to_datetime('2006-01-01')
```

```python
complaints_subset = complaints_sorted[morethandate]
```

```python

complaints_subset['LONGLAT']=complaints_subset['Longitude'].astype(str)+" "+complaints_subset['Latitude'].astype(str)
complaints_subset
```

```python
arrests_sorted = arrests.sort_values(by=['ARREST_DATE'])
arrests_sorted
```

```python
arrests_sorted['ARREST_DATE']=pd.to_datetime(arrests_sorted['ARREST_DATE'], errors='coerce')
arrests_sorted['LONGLAT']=arrests_sorted['Longitude'].astype(str)+" "+arrests_sorted['Latitude'].astype(str)
arrests_sorted
```

```python
arrests_sorted.columns = arrests_sorted.columns.to_series().replace({'ARREST_DATE':'DATE'})

complaints_subset.columns = complaints_subset.columns.to_series().replace({'CMPLNT_FR_DT':'DATE'})
```

```python
complaints_subset
```

# Complaints that result in arrests

Using date of the complaint / arrest and the longitude and latitude

```python
print(arrests_sorted.columns)
print(complaints_subset.columns)

arrests_sorted.columns = arrests_sorted.columns.to_series().replace({'PERP_SEX':'SUSP_SEX','PERP_RACE':"SUSP_RACE","AGE_GROUP":"SUSP_AGE_GROUP"})
```

```python
print(arrests_sorted['SUSP_SEX'].unique())
print(arrests_sorted['SUSP_RACE'].unique())
print(arrests_sorted['SUSP_AGE_GROUP'].unique())

print(arrests_sorted['X_COORD_CD'])
print(arrests_sorted['OFNS_DESC'])
print(arrests_sorted['LAW_CAT_CD'].unique())
```

```python
print(complaints_subset['SUSP_SEX'].unique())
print(complaints_subset['SUSP_RACE'].unique())
print(complaints_subset['SUSP_AGE_GROUP'].unique())
print(complaints_subset['X_COORD_CD'])
print(complaints_subset['OFNS_DESC'])
print(complaints_subset['LAW_CAT_CD'].unique())
```

```python
complaints_subset['LAW_CAT_CD']=complaints_subset['LAW_CAT_CD'].str[0]
```

```python
arrest_complaints = pd.merge(arrests_sorted, complaints_subset, 
                             on=['DATE','LONGLAT','SUSP_SEX','SUSP_RACE','X_COORD_CD','Y_COORD_CD','JURISDICTION_CODE','LAW_CAT_CD'], how='inner')
```
